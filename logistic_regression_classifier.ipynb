```python
from load_mnist import * 
import numpy as np 

#load training and testing dataset
X_train, y_train = load_mnist('training')
X_test, y_test = load_mnist('testing')

#reshape 28*28 into single row
X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_test = np.reshape(X_test, (X_test.shape[0], -1))

print(X_train.shape)
print(X_test.shape)
```

```python
def softmax_activation(dot_matrix):
    dot_matrix = dot_matrix - np.max(dot_matrix)
    return np.argmax((np.exp(dot_matrix.T) / np.sum(np.exp(dot_matrix))).T)
```

```python
#from sklearn.metrics import mean_squared_error
def network_training(X_batch, y_batch, W, b):
    for each_itr in range(100):
        dot_matrix = np.dot(X_batch, W) + b
        #print(dot_matrix)
        softmax = softmax_activation(dot_matrix)
        #print(softmax)
        mse = y_batch - softmax.T  #mean_squared_error(y_batch, softmax.T)
        #print(mse)
        W += np.dot(X_batch.T, mse)
        return W
```

```python
#initialize weights
np.random.seed(1)

W = np.random.rand(X_train.shape[1])
b = np.zeros((1,1))
batch_size = 20

for i in range(len(X_train)):
    if(i % batch_size == 0):
        X_batch = X_train[i:i+batch_size]
        y_batch = y_train[i:i+batch_size]
    
    W = network_training(X_batch, y_batch, W, b)
```

```python
# Test dataset prediction
y_actual = softmax_activation(np.dot(X_test,W)+b)
x, _ = np.unique(y_actual,return_counts=True)
accuracy = ((len(y_test) - x[0])* 100) / len(y_test)
print(accuracy)
```
